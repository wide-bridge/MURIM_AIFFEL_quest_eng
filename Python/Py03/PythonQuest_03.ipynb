[프로그램 개요]

사용자로부터 두 개의 정수와 **연산자(+, -, *, /, **)**를 입력받아 계산 결과를 출력하는 프로그램을 작성하세요.

[상세 조건]

정수 입력 예외 처리: 사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류

*   항목 추가
*   항목 추가

메시지를 출력해야 합니다.

입력 반복: 정수가 입력될 때까지 **"잘못된 입력입니다. 정수를 입력해주세요."**를 출력하며 다시 입력을 시도합니다.

0으로 나누기 제한: 나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.

지원하지 않는 연산자: 사용자가 목록에 없는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.

math 모듈 활용: 제곱 연산(**)은 파이썬 내장 math 모듈을 사용하여 지원합니다.

반복 여부 확인: 계산이 완료된 후 계산기를 통해 계속 계산을 진행할 것인지 입력을 받습니다.

```python
import math

def get_integer(prompt):
    """정수를 입력받을 때까지 반복하는 함수"""
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

def get_operator():
    """올바른 연산자를 입력받을 때까지 반복하는 함수"""
    valid_operators = ['+', '-', '*', '/', '**']
    while True:
        op = input(f"연산자를 입력하세요 ({', '.join(valid_operators)}): ")
        if op in valid_operators:
            return op
        else:
            print(f"잘못된 입력입니다. {valid_operators} 중 하나를 입력해주세요.")

def run_calculator():
    print("=== 파이썬 예외 처리 계산기 (검증 강화 버전) ===")

    while True:
        # 1. 정수 입력 받기 (올바른 값일 때까지 반복)
        num1 = get_integer("\n첫 번째 정수를 입력하세요: ")
        num2 = get_integer("두 번째 정수를 입력하세요: ")

        # 2. 연산자 입력 받기 (올바른 값일 때까지 반복)
        operator = get_operator()

        # 3. 연산 수행 및 예외 처리
        try:
            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                result = num1 / num2
            elif operator == '**':
                # math 모듈을 사용하여 제곱 연산 수행
                result = math.pow(num1, num2)

            print(f">>> 계산 결과: {result}")

        except ZeroDivisionError:
            print("오류: 0으로 나눌 수 없습니다.")
        except OverflowError:
            print("오류: 계산 결과가 너무 커서 표시할 수 없습니다.")

        # 4. 계속 계산할지 여부 확인
        cont = input("\n계속 계산하시겠습니까? (종료: n / 계속: 아무 키): ").lower()
        if cont == 'n':
            print("프로그램을 종료합니다.")
            break
        print("-" * 35)

if __name__ == "__main__":
    run_calculator()
```

```python
#2안

import math

def get_integer(prompt):
    """정수를 입력받을 때까지 반복하는 함수 (ValueError 처리)"""
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

def run_calculator():
    print("=== 파이썬 예외 처리 계산기 ===")

    while True:
        # 1. 정수 입력 받기
        num1 = get_integer("첫 번째 정수를 입력하세요: ")
        num2 = get_integer("두 번째 정수를 입력하세요: ")

        # 2. 연산자 입력 받기
        operator = input("연산자를 입력하세요 (+, -, *, /, **): ")

        # 3. 연산 수행 및 예외 처리
        try:
            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                # ZeroDivisionError 처리
                result = num1 / num2
            elif operator == '**':
                # math 모듈을 사용하여 제곱 연산 수행
                result = math.pow(num1, num2)
            else:
                print(f"오류: '{operator}'는 지원하지 않는 연산자입니다.")
                result = None

            if result is not None:
                print(f"계산 결과: {result}")

        except ZeroDivisionError:
            print("오류: 0으로 나눌 수 없습니다. (ZeroDivisionError)")
        except Exception as e:
            print(f"예상치 못한 오류가 발생했습니다: {e}")

        # 4. 계속 계산할지 여부 확인
        cont = input("\n계속 계산하시겠습니까? (종료하려면 'n' 입력, 계속하려면 아무 키나 입력): ").lower()
        if cont == 'n':
            print("프로그램을 종료합니다.")
            break
        print("-" * 30)

if __name__ == "__main__":
    run_calculator()
```
